```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score, accuracy_score, classification_report
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder

# path = 'Data/processed_dataset.csv'
# df = pd.read_csv(path)

# print(df.head())
# print(df.shape)
```

```python
flight_df = pd.read_csv('./Data/initial_dataset.csv')
```

```python
flight_df
```

```python
columns_to_drop = ["TAIL_NUM"]
```

```python
flight_df.drop(columns=columns_to_drop, inplace=True)
```

```python
'''
Label encoding
'''
le = LabelEncoder()
```

```python
for col in flight_df.columns:
    if type(flight_df[col][0])==str:
        flight_df[col] = le.fit_transform(flight_df[col])
```

```python
flight_df
```

```python
def process(l):
    if l<=0:
        return 0
    else:
        return 1
```

```python
flight_df['DEP_DELAY']=flight_df['DEP_DELAY'].apply(lambda l: process(l))
```

```python
flight_df
```

```python
X = flight_df.drop(columns='DEP_DELAY')
Y = flight_df.DEP_DELAY
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.4, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
svm_model = SVC(kernel='poly', C=2.5, random_state=42)
svm_model.fit(X_train_scaled, y_train)
```

```python
y_pred = svm_model.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print(f'Classification Report:\n', classification_rep)
```
