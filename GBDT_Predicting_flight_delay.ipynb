```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import r2_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score, accuracy_score, classification_report, mean_absolute_error, mean_squared_error, r2_score, confusion_matrix
import matplotlib.pyplot as plt
import numpy as np
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
```

```python
file_path = "C:/Users/delni/OneDrive/Documents/Courses/AML/Project/data/M1_final.csv"
data = pd.read_csv(file_path)
data
```

```python
columns_to_drop = ["TAIL_NUM"]
data.drop(columns=columns_to_drop, inplace=True)
```

```python
le = LabelEncoder()
for col in data.columns:
    if type(data[col][0])==str:
        data[col] = le.fit_transform(data[col])
```

```python
# Fill missing values if any (example: using mode for categorical and median for numerical)
for column in data.columns:
    if data[column].dtype == 'object':
        data[column] = data[column].fillna(data[column].mode()[0])  #Filling Missing Values for Categorical Columns
    else:
        data[column] = data[column].fillna(data[column].median())  #Filling Missing Values for Numeric Columns      
```

```python
data
data['DEP_DELAY'].plot(figsize=(10, 6), label='departure delay')
data.describe()
```

```python
# Split the data into features and targets
X = data
y = data.DEP_DELAY
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
y_train
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled.shape
```

```python
X_test_scaled.shape
```

```python
X_train_scaled = np.delete(X_train_scaled, 5, axis=1)
X_train_scaled.shape
```

```python
X_test_scaled = np.delete(X_test_scaled, 5, axis=1)
X_test_scaled.shape
```

```python
# Initialize and train the GBDT model
gbdt_regressor = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
gbdt_regressor.fit(X_train_scaled, y_train)
```

```python
# Make predictions and evaluate the model
y_pred = gbdt_regressor.predict(X_test_scaled)
```

```python
(((y_pred-y_test)**2).mean())**0.5
```

```python
mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)  
print(f'Mean Absolute Error (MAE): {mae:.2f}')
print(f'Root Mean Squar Error (RMS): {rmse:.2f}')
print(f'Coefficient of Determination (R^2): {r2:.2f}')
```
